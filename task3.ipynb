# Task3: Visualization Using Histogram and Bar Charts

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=pd.read_csv(r"C:\Users\HP\Desktop\Technohack data analytics internship\Task3- Visualization using histogram\Iris.csv")
```

```python
iris.head()
```

```python
plt.figure(figsize=(10,7))
x=iris.Species
y=iris.SepalLengthCm
plt.bar(x,y,color="pink")
plt.title("Species by Sepal lenght")
plt.xlabel("Species")
plt.ylabel("SepalLength in cm")
plt.show()
```

```python
plt.figure(figsize=(10,7))
x=iris.Species
y=iris.SepalWidthCm
plt.bar(x,y,color="black")
plt.title("Species by Sepal lenght")
plt.xlabel("Species")
plt.ylabel("Sepalwidth in cm")
plt.show()
```

```python
plt.figure(figsize=(10,7))
x=iris.Species
y=iris.PetalLengthCm
plt.bar(x,y,color="yellow")
plt.title("Species by Sepal lenght")
plt.xlabel("Species")
plt.ylabel("PetalLength in cm")
plt.show()
```

```python
plt.figure(figsize=(10,7))
x=iris.Species
y=iris.PetalWidthCm
plt.bar(x,y,color="orange")
plt.title("Species by Sepal lenght")
plt.xlabel("Species")
plt.ylabel("Petalwidth in cm")
plt.show()
```

# Histogram

```python
plt.figure(figsize = (10, 7))
x = iris["SepalLengthCm"]
  
plt.hist(x, bins = 20, color = "green")
plt.title("Sepal Length in cm")
plt.xlabel("Sepal_Length_cm")
plt.ylabel("Count")
```

```python

plt.figure(figsize = (10, 7))
x = iris.SepalWidthCm
  
plt.hist(x, bins = 20, color = "yellow")
plt.title("Sepal Width in cm")
plt.xlabel("Sepal_Width_cm")
plt.ylabel("Count")
  
plt.show()
```

```python


plt.figure(figsize = (10, 7))
x = iris.PetalLengthCm
  
plt.hist(x, bins = 20, color = "Blue")
plt.title("Petal Length in cm")
plt.xlabel("Petal_Length_cm")
plt.ylabel("Count")
  
plt.show()
```

```python

plt.figure(figsize = (10, 7))
x = iris.PetalWidthCm
  
plt.hist(x, bins = 20, color = "orange")
plt.title("Petal Width in cm")
plt.xlabel("Petal_Width_cm")
plt.ylabel("Count")
  
plt.show()
```
