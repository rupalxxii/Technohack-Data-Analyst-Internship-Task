# Calculate Summary statistic 

```python
import pandas as pd
```

```python
test =pd.read_csv(r"C:\Users\HP\Desktop\Technohack data analytics internship\Task2-Calculate Summary statistics\test.csv")
```

```python
test.head()
```

# Aggregating statistics

What is the average age of the Titanic passengers?

```python
test["Age"].mean()
```

What is the median age and ticket fare price of the Titanic passengers?

```python
test[["Age","Fare"]].median()
```

The aggregating statistic can be calculated for multiple columns at the same time.

```python
test.describe()
```

```python
test[["Age","Fare"]].describe()
```

```python
test
```

Instead of the predefined statistics, specific combinations of aggregating statistics for given columns can be defined using the DataFrame.agg() method:

```python
test.agg(
    {
        "Age": ["min", "max", "median", "skew"],
        "Fare": ["min", "max", "median", "mean"],
    }
)
```

# Aggregating statistics grouped by category

What is the average age for male versus female Titanic passengers?

```python
test[["Sex", "Age"]].groupby("Sex").mean()
```

```python
test.groupby("Sex").mean(numeric_only=True)
```

```python
test.groupby("Sex").mean(numeric_only=True)
```

What is the mean ticket fare price for each of the sex and cabin class combinations?

```python
test.groupby(["Sex", "Pclass"])["Fare"].mean()
```

# Count number of records by category


What is the number of passengers in each of the cabin classes?

```python
test["Pclass"].value_counts()
```

```python
test.groupby("Pclass")["Pclass"].count()
```
